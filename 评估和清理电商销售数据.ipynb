# 项目：评估和清理英国电商公司销售数据

## 分析目标

此数据分析的目的是，根据市场销售数据，挖掘畅销产品，以便制定更有效的市场策略来提升营收。

本实战项目的目的在于练习评估数据干净和整洁度，并且基于评估结果，对数据进行清洗，从而得到可供下一步分析的数据。

## 简介

原始数据集记录了一家英国在线零售公司在2010年12月1日至2011年12月9日期间的所有交易情况，涵盖了该公司在全球不同国家和地区的业务数据。该公司主要销售覆盖各个场景的礼品，包括但不限于生日礼品、结婚纪念品、圣诞礼品等等。该公司的客户群体主要包括批发商和个人消费者，其中批发商占据了相当大的比例。

数据每列的含义如下：
- `InvoiceNo`: 发票号码。6位数，作为交易的唯一标识符。如果这个代码以字母“c”开头，表示这笔交易被取消。
- `StockCode`: 产品代码。5位数，作为产品的唯一标识符。
- `Description`: 产品名称。
- `Quantity`: 产品在交易中的数量。
- `InvoiceDate`: 发票日期和时间。交易发生的日期和时间。
- `UnitPrice`: 单价。价格单位为英镑（£）。
- `CustomerID`: 客户编号。5位数，作为客户的唯一标识符。
- `Country`: 国家名称。客户所居住的国家的名称。

## 清理数据

根据前面的评估部分得到的结论，需要对数据清理包括：
把InvoiceDate变量数据类型转换为日期时间
把CustomerID变量数据转化为字符串类型
把Desciption变量缺失值删除
把Country变量值USA替换为United States;UK U.K 替换为United Kindom
把quntity变量值为负数的删除
把unitprice变量值为负数的观察值删除


 为了区分清理后的数据和原始数据，创建新变量clean_df，让他为df复制出来的副本，之后所有的操作均再clean_df进行

```python
clean_df=df.copy()
clean_df
```

```python
clean_df["InvoiceDate"]=pd.to_datetime(clean_df["InvoiceDate"])#时间格式更改pd.to_datetime 然后进行赋值
```

```python
#df["CustomerID"]=df["CustomerID"].astype("int") #这是错误的,customerID 里面包含空值
#df.info() 查看后，发现确实已经更改，但是我们需要删除0
```

```python
clean_df["CustomerID"]=clean_df["CustomerID"].astype(str)
clean_df#发现CID 含有 0 用str的slice方法
```

```python
clean_df["CustomerID"].str.slice(0,-2)
clean_df["CustomerID"]=df["CustomerID"].str.slice(0,-2)
```

```python
clean_df
```

```python
clean_df.dropna(subset="Description",inplace=True)
```

```python
clean_df["Description"].isnull().sum()
```

替换值 ：replace 


```python
clean_df=clean_df.replace(["UK","U.K."],"United Kingdom")

```

```python
clean_df["Country"]=clean_df["Country"].replace("USA","United States")
```

```python
clean_df["Country"].value_counts()
```

```python
print(len(clean_df[clean_df["Country"]=="USA"]))#检查 country为USA行数的长度
print(len(clean_df[clean_df["Country"]=="UK"]))
print(len(clean_df[clean_df["Country"]=="U.K"]))
```

Quantity<0 & Unit Price<0 的值全部删除 思路：保留 数量大于0以及 单价大于0的行数

```python
clean_df["Quantity"]=clean_df["Quantity"][clean_df["Quantity"]>0]
clean_df["UnitPrice"]=clean_df["UnitPrice"][clean_df["UnitPrice"]>0]
```

```python
print(len(clean_df[clean_df["Quantity"]<0]))
print(len(clean_df[clean_df["UnitPrice"]<0]))
```
